```python
# YOUR CODE HERE
numeric_df = df.select_dtypes(include=['number'])
columns_to_plot = ['year', 'co2', 'comb']
plt.figure(figsize=(14, 10))
for i, col in enumerate(columns_to_plot, 1):
    plt.subplot(2, 3, i)
    plt.hist(df[col], bins=20, color='skyblue', edgecolor='black')
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.savefig('/content/combined_histograms.png')
files.download('/content/combined_histograms.png')
plt.show()
plt.close()
```

interpretation of the gaphs

year:
This chart shows how many cars in the dataset were from each year. From the bars, we can see that the data is evenly spread from 2013 to 2018, meaning there are almost the same number of cars for each of those years. This tells us the data doesnt favor one year over another  it's balanced.

CO2:
This chart shows how much carbon dioxide (CO) the cars produce. Most cars in the dataset produce around 300 to 400 grams of CO. Fewer cars produce either very low or very high amounts. So the chart looks like a hill  with most cars in the middle and few on the edges. This means that average emissions are most common, and extremely clean or extremely polluting cars are rare in this dataset.

Comnb:
This chart shows how many miles per gallon (mpg) the cars can drive on average. Most cars have a fuel economy of around 20 to 30 mpg, which means they drive that many miles per gallon of fuel. Very few cars go much higher or lower. So, most cars in this dataset are average fuel savers, and only a few are very efficient or very inefficient.


```python

corr = numeric_df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap of Numeric Features', fontsize=10)
plt.tight_layout()
plt.savefig("heat_map.png", dpi=300)
plt.show()
plt.close()

plt.figure(figsize=(8, 5))
plt.hist(df['comb'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Combined Fuel Efficiency (MPG)')
plt.xlabel('MPG (Combined)')
plt.ylabel('Number of Vehicles')
plt.grid(True)
plt.tight_layout()
plt.savefig("histogram_fuel_efficiency.png", dpi=300)
plt.close()
from google.colab import files
files.download("heat_map.png")

```
